```python
from PIL import Image
import numpy as np
import os
```

```python
path = "ArkNights\\Texas.png"
im01 = Image.open(path)
arr1 = np.array(im01)

for i in range(5):
    for j in range(5):
        arr1[arr1.shape[0]-i-1][arr1.shape[1] - j-1] = [255,255,255,1]
        arr1[arr1.shape[0]-i-1][j] = [255,255,255,1]
        arr1[i][arr1.shape[1] - j-1] = [255,255,255,1]
        arr1[i][j] = [255,255,255,1]
im2 = Image.fromarray(arr1)
im2.save(path)
```
